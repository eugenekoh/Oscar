```python
from oscar.modeling.modeling_bert import BertForPersonalityImageCaptioning
from transformers.pytorch_transformers import BertTokenizer, BertConfig
import os
from tensorboardX import SummaryWriter
```

```python
os.chdir('/home/eugene/workspace/Oscar')
```

```python
checkpoint = './experiments/inject/checkpoint-12-9000'
config = BertConfig.from_pretrained(checkpoint)
model = BertForPersonalityImageCaptioning.from_pretrained(checkpoint, config=config)
```

# Visualize Embeddings

```python
writer = SummaryWriter('runs/visualize_embedding')
```

```python
with open('./datasets/personality_captions/personalities.txt') as f:
    words = list(x.replace('\n', '') for x in f.readlines())
```

```python
weights  = model.personality_embedding.weight
writer.add_embedding(weights, metadata=words, tag=f'personality embedding', global_step=1)
```
